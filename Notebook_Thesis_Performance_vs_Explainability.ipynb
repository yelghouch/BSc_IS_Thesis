```python
x_1000=my_dataset[my_dataset['isFraud']==1]
x_1000=x_1000.head(1000)
y_1000=my_dataset[my_dataset['isFraud']==0]
y_1000=y_1000.head(1000)
frames = [x_1000,y_1000]
setA = pd.concat(frames)
numerical = ['type','amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
pd.get_dummies(setA['type'], prefix='type')

dfA = pd.concat([setA,pd.get_dummies(setA['type'], prefix='type')],axis=1)

dfA.drop(['type'],axis=1, inplace=True)
dfA.drop(['step'],axis=1, inplace=True)
dfA.drop(['nameOrig'],axis=1, inplace=True)
dfA.drop(['nameDest'],axis=1, inplace=True)
```

```python
x_500000=my_dataset[my_dataset['isFraud']==1]
x_500000=x_500000.head(500000)
y_500000=my_dataset[my_dataset['isFraud']==0]
y_500000=y_500000.head(500000)
frames = [x_500000,y_500000]
setB = pd.concat(frames)
numerical = ['type','amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
pd.get_dummies(setB['type'], prefix='type')

dfB = pd.concat([setB,pd.get_dummies(setB['type'], prefix='type')],axis=1)

dfB.drop(['type'],axis=1, inplace=True)
dfB.drop(['step'],axis=1, inplace=True)
dfB.drop(['nameOrig'],axis=1, inplace=True)
dfB.drop(['nameDest'],axis=1, inplace=True)

```

```python
x_646=my_dataset[my_dataset['isFraud']==1]
x_646=x_646.head(646)
y_500000=my_dataset[my_dataset['isFraud']==0]
y_500000=y_500000.head(500000)
frames = [x_646,y_500000]
setC = pd.concat(frames)
numerical = ['type','amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
pd.get_dummies(setC['type'], prefix='type')

dfC = pd.concat([setC,pd.get_dummies(setC['type'], prefix='type')],axis=1)

dfC.drop(['type'],axis=1, inplace=True)
dfC.drop(['step'],axis=1, inplace=True)
dfC.drop(['nameOrig'],axis=1, inplace=True)
dfC.drop(['nameDest'],axis=1, inplace=True)
```

```python
features = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'type_CASH_IN', 'type_CASH_OUT', 'type_DEBIT', 'type_PAYMENT', 'type_TRANSFER']
X = dfA[features]
y = dfA['isFraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state =1)
```

# Performance

## k-NN

```python
knn = KNeighborsClassifier()
knn_scores = cross_val_score(knn, X_train,y_train, cv=10, scoring='precision')
print(knn_scores)
knn.fit(X_train, y_train)
```

```python
predictions1 = knn.predict(X_test)
```

```python
score1 = knn.score(X_test, y_test)
print(score1)
```

```python
cm1 = metrics.confusion_matrix(y_test, predictions1)
print(cm1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions1))
```

## DT

```python
dt = DecisionTreeClassifier()
dt_scores = cross_val_score(dt, X_train,y_train, cv=10, scoring='precision')
print(dt_scores)
dt.fit(X_train, y_train)
```

```python
predictions2 = dt.predict(X_test)
```

```python
score1 = dt.score(X_test, y_test)
print(score1)
```

```python
cm2 = metrics.confusion_matrix(y_test, predictions2)
print(cm2)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions2))
```

## MLP

```python
mlp = MLPClassifier()
mlp_scores = cross_val_score(mlp, X_train,y_train, cv=10, scoring='precision')
print(mlp_scores)
mlp.fit(X_train, y_train)
```

```python
predictions3 = mlp.predict(X_test)
```

```python
score1 = mlp.score(X_test, y_test)
print(score1)
```

```python
cm3 = metrics.confusion_matrix(y_test, predictions3)
print(cm3)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions3))
```
